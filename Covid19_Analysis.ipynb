# **Covid19 Analysis** 

### **Importing the modules**

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
```

### **Importing covid19 dataset**

```python
covid_data=pd.read_csv("covid19_Confirmed_dataset.csv")
covid_data.head()
```

### **Deleteing the useless columns**

```python
covid_data.drop(["Lat","Long"],axis=1,inplace=True)
covid_data.head()
```

### **Aggregating the rows by the country**

```python
covid_aggregating=covid_data.groupby("Country/Region").sum()
covid_aggregating.head()
```

```python
covid_aggregating.shape
```

### **Visualizing data related to a country for example China**

```python
covid_aggregating.loc["China"].plot()
covid_aggregating.loc["Italy"].plot()
covid_aggregating.loc["Spain"].plot()
plt.legend()
```

### **Calculating a good measure**
we need to find a good measure reperestend as a number, describing the spread of the virus in a country.

```python
covid_aggregating.loc['China'].plot()
```

```python
covid_aggregating.loc['China'][:3].plot()
```

### **caculating the first derivative of the curve**

```python
covid_aggregating.loc['China'].diff().plot()
```

### **find maximum infection rate for all of the countries.**

```python
countries=list(covid_aggregating.index)
max_infection_rates=[]
for c in countries:
    max_infection_rates.append(covid_aggregating.loc[c].diff().max())
covid_aggregating["max_infection_rates"]=max_infection_rates


covid_aggregating.head()
```

### **create a new dataframe with only needed column**

```python
data=pd.DataFrame(covid_aggregating["max_infection_rates"])
data.head()
```

### **Importing Happiness Dataset**

```python
happiness=pd.read_csv("worldwide_happiness_report.csv")
happiness.head()
```

### **Droping the useless columns**

```python
cols=["Overall rank","Score","Generosity","Perceptions of corruption"]

happiness.drop(cols,axis=1,inplace=True)

happiness.head()
```

### **changing the indices of the dataframe**

```python
happiness.set_index("Country or region",inplace=True)
happiness.head()
```

### **Joining two dataset we have prepared**

Corona Dataset :

```python
data.head()
```

wolrd happiness report Dataset :

```python
happiness.head()
```

```python
final=data.join(happiness,how="inner")
final.head()
```
